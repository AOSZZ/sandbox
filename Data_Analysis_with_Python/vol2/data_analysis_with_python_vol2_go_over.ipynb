### excelを読み込むと欠損値が多いため、分析しやすいデータに直したい

**```NaN``` を単純に除去すると、データ構造(例えば、市町村区とそれに関連する統計値)がずれてしまう...**  
**なので、地道に加工していく必要がある**　　

````python
excel_df.dropna(how='all')
````

**行方向に、すべての列が```NaN``` である場合、その行を除去してみる、**  
**一行分除くことができたが、一方で、他全ての行は、何かしら文字列や値が入っていることを意味する**

````python
excel_df.dropna(axis=1, how='all')
````

**今度は列方向に、すべての行が```NaN``` である場合、その列を除去してみる、**  
**２列分除くことができたが、一方で、他全ての列は、何かしら文字列や値が入っていることを意味する**

````python
df = excel_df.dropna(how='all')
df = df.dropna(axis=1, how='all')
````

````python
df
````

**少し、全体がみやすくなった？**  
**excel に限らず、何かしらデータファイルを読み込んだとき、```dropna``` 引数 ```how='all'```を試すといいかもしれない**

**ここで、行と列が除去されたためにインデックスに抜けができてしまった...**  
**```reset_index``` 引数```drop=True``` でインデックスを振り直す**

````python
df_r = df.reset_index(drop=True)
````

````python
df_r
````
